#**The plt.plot() function**

The most fundamental command to matplot lib. There are so many parameters to this command, the options are endless!

This pushes the data into the frontend that you and I see, that is the graph canvas. What goes up there and how it goes, we decide.

```python
plt.plot([1,2,3],[4,5,6], label="Label", color = 'red', linewidth = 5, mfc = "blue",
         marker= "8", markersize = 15, markeredgecolor= 'yellow', linestyle = "-.")

plt.show() #Shows the plot to the user.
```

***So when we run this cell it should plot (1,2) ; (2,4) ; (3,6) as the points, thereby yielding a straight line.***

Lets try and understand this code bit by bit



`label` : Allows you to set a label to refer to in the legend

`linewidth` : Allows you to change the thickness of the plot line

`marker` : marks all the points you have plotted with certain strings you can choose.  

`markeredgecolor` : Lets you choose the color of the outer perimeter of the marker

`markerfillcolor` : Lets you choose the fill color of the marker (AKA mfc)

`linestyle` : allows you to choose what kinds of lines you want.


A list of all the markers that MatPlotLib supports.

We can speed up the process by using the shorthand notation which goes like **color | marker | line**. Lets see an example of this :

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y, 'r.--', label="2x" )
plt.legend()
plt.show()
```

**Basic Commands:**

```python
# lets customize a graph

plt.xlabel("X") #sets the labelling for both the X and the Y axes
plt.ylabel("Y")

plt.title("Single Graph", fontdict = {"fontname" : "Verdana", "fontsize" : 20}) #colab doesn't contain the fonts so it defaults to one.

plt.xticks([0,1,2,3,4,5]) #Gives the ticks data in X and Y axes
plt.yticks([0,2,4,6,8,10])


x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]

plt.plot(x, y, 'r.-', label="2x", markersize = 10)
plt.legend() #runs a background function that will print out the legend you see in the top left.
plt.show()
```

##**Plotting a second line in the same graph canvas**

```python
plt.xlabel("X") #sets the labelling for both the X and the Y axes
plt.ylabel("Y")

plt.title("Single Graph",fontdict = {"fontname" : "Roboto", "fontsize" : 20}) #colab doesn't contain the fonts so it defaults to one.

x = [0,1,2,3,4,5, 6]
y = [0,2,4,6,8,10,12]


plt.plot(x, y, 'r.--', label="2x", markersize = 10) #plotting the graph, for which we set up the variables earlier

x1 = list(np.arange(0,8,2))
y1 = np.square(x1) #useful numpy command to square all the elements of the array

xtick = x1[::2] #using list slicing to generate step values.
ytick = y1[::2]


plt.plot(x1,y1,'g.--',label = "X^2", markeredgecolor = 'b', mfc = 'r', markersize = 10)
plt.legend()
plt.show()
```

Notice how the graph scales with the values we input.

#Bar Graphs


It is a pretty simple graph to work with, and is useful in comparitive datatypes.

```python
#initialising a bar graph

x = np.array(["A", "B", "C", "D"])
y = np.array([1, 2, 3, 4])

plt.bar(x,y)
plt.show()
```

Once again, all the customisation options persist here also.

Use `.barh` instead of `.bar` to get a horizontal bar graph.

```python
#initialising a horizontal bar graph
x = np.array(["A", "B", "C", "D"])
y = np.array([1, 2, 3, 4])

plt.barh(x,y)
plt.show()
```

We can customise individual bars of the bar graphs by using patterns.

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([1, 2, 3, 4])

bars = plt.barh(x,y)

bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch("*")
bars[3].set_hatch("x")
plt.figure(figsize=(6,4))
plt. show()
```
