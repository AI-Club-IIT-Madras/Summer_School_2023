```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
img = cv2.imread('coldplay.jpeg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_detector.detectMultiScale(gray, 1.3, 5)
# 1.3 indicates scaleFactor which is a parameter that specifies how much the image size is reduced during the detection pass
# 5 is minNeighbours which is a parameter which specifies how many neighbours a candidate rectangle should have to retain it
for (x, y, w, h) in faces:
  cv2.rectangle(img, (x,y), (x+w, y+h), (0, 0, 255), 3)

cv2_imshow(img)
```
