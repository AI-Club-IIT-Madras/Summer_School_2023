Numpy is a very important, useful and famous library as it offers multi-dimensional array capabilities. Lists store values, numpy arrays have built in functions that we can use to manipulate the data en masse.

Now let us turn our attention to using Numpy to manipulate the data. Remember that 'rows' contains our data and 'columns' contains the headings.

Operation to convert an interable array A to a numpy object \\
A_array = np.array(A) \\
A.ndim returns dimensions \\
A.shape returns shape

```python
import numpy as np

# The data is in Python lists. To use NumPy we need to convert them into NumPy arrays.
# To keep it simple for analysis, let's use just some of this data.

rows_array = np.array(rows)  # **Convert rows to a NumPy object**

print("The array: ",rows_array )  # **Print the array**
print("Dimension: \n", rows_array.ndim )  # **Print the number of dimensions**
print("Shape: \n",rows_array.shape )  # **Print the shape of the array**
```

**Slicing** \\
If 2 dimension , the syntax is A[range_1, range_2] \\
range = [start:end+1] \\
remember columns are indexed from 1 not 0 \\
A.astype(dtype) casts array A to a datatype \\
np.int64 = 64 bit integer datatype



```python
# Get the numerical columns in the dataset, calories and protein.
calories = rows_array[:,3 ]  # **Get every row, with column number being index for calories**
protein = rows_array[:,4]  # **Same but for protein**

# Cast calories and protein as 64-bit integers.
calories = calories.astype(np.int64)  # **Cast calories as a 64-bit integer**
protein = protein.astype(np.int64)  # **Cast protein as a 64-bit integer**

print("Calories array - \n", calories)
print("Protein array - \n", protein)
```

```python
# lets find out more about the calories array
print("Datatype of the calories array-", calories.dtype) #prints the datatype of the array

print("Item size -", calories.itemsize) #prints the size of each item, we took int8, its 1 byte per item

print("Net array size -", calories.size) #prints net size of the array.
```

```python
# how about if we want a particular element?

print(rows_array[1,0])
```

Functional operations on np_objects \\
np.square(A) = returns an array with same dimensions as A with it squaring every element within it \\
np.sin(A) = sin function on every element \\
np.power(A, n) = raises every element to power n

```python
# Now for operations. First, let's square all elements in the second column and find the sine value of elements in the third column.

calorie_squares = np.square(calories)  # **Use squaring**
protein_sines = np.sin(protein)   # **Use sine function**
calorie_squares_alternate = np.power(calories,2)  # **Use power function**

print("Calories ^2: \n", calorie_squares)
print("sin(protein): \n", protein_sines)
print("Calories^2 alternate: \n", calorie_squares_alternate)
```

np.add(A,B) returns an array that contains every element to be the sum of it's respective analgous elements in A and B \\
np.subtract(A,B) subtracts B from A \\
np.multiply(A,A) multiplies and returns element wise

```python
# lets see how the add, subtract and multiply commands work

calories_plus_protein =np.add(calories,protein)
calories_minus_protein =np.subtract(calories,protein)
calories_times_calories =np.multiply(calories,calories)

print("addition -\n ", calories_plus_protein)
print("subtraction -\n ", calories_minus_protein)
print("multiplication -\n ", calories_times_calories) # compare this with calories^2 in the earlier cell
```

Shape manipulation is an important part of numpy usage. Lets see how we can do that:

**How does Ravel work** \\
ravel.png

```python
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[1,2,3],[4,5,6]])
```

```python
# lets reshape the vertical stack
print("unravelled -\n", a.ravel())

# reshape command-
reshaped = b.reshape(3,2)
print("after using reshape-\n", reshaped)
print("original array-\n", b)

#resize command:
b.resize(3,2)
print("original array has changed-\n", b)
```

An important difference between reshape and resize is highlighted here. Reshape return a new array leaving the old one unchanged while resize changes the original array.
